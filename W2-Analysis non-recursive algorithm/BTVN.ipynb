**1. Big O**

Giả sử f(n) và g(n) là các hàm thực không âm của đối số nguyên không âm n. Ta nói "g(n) là O của f(n)" và viết là: g(n) = O(f(n)) nếu tồn tại các hằng số dương C và n0 sao cho g(n) <= C * f(n) với mọi n >= n0

**Cách tính**

Độ phức tạp tính toán của giải thuật: O(f(n))

• Việc xác định độ phức tạp tính toán của giải thuật trong thực tế có thể tính bằng một số quy tắc đơn giản sau:

– Quy tắc bỏ hằng số:

        T(n) = O(c.f(n)) = O(f(n)) với c là một hằng số dương

– Quy tắc lấy max:

        T(n) = O(f(n)+ g(n)) = O(max(f(n), g(n)))

– Quy tắc cộng:

        T1(n) = O(f(n))                     T2(n) = O(g(n))

        T1(n) + T2(n) = O(f(n) + g(n))

– Quy tắc nhân:

        Đoạn chương trình có thời gian thực hiện T(n)=O(f(n))

        Nếu thực hiện k(n) lần đoạn chương trình với k(n) = O(g(n)) thì độ phức tạp sẽ là O(g(n).f(n))

**Run time**

Mặc dù có thể tính toán thời gian chạy chính xác hơn có thể giúp ích khi hiểu các thuật toán cụ thể, nhưng nhiều khi bạn có thể phân loại chương trình bằng một trong các cách sau (được sắp xếp từ nhanh nhất đến dài nhất). Gọi n là kích thước đầu vào của chúng ta:


*   (1) được gọi là **thời gian không đổi**. Ở đây, kích thước của n không ảnh hưởng đến tốc độ.
*   (log_c (n)) được gọi là **thời gian logarit**. Ở đây, thời gian chạy của chúng tôi thay đổi đối với log-base-c của n)

*   (n) được gọi là **thời gian tuyến tính**. Ở đây, thời gian chạy của chúng ta thay đổi trực tiếp đối với n.

*   (n log (n)) được gọi là **thời gian siêu tuyến tính**. Ở đây, thời gian chạy của chúng ta chậm hơn so với tuyến tính, nhưng không đủ để trở thành đa thức.

*   (n ^ c) được gọi là **thời gian đa thức**. Ở đây, thời gian chạy của chúng ta đã chậm lại và phụ thuộc vào độ thứ c của n.

*   (c ^ n) được gọi là **thời gian hàm mũ**. Ở đây, thời gian chạy của chúng tôi đã trở nên tỷ lệ thuận với một giá trị, c, được nâng lên kích thước đầu vào của chúng tôi. Rất chậm và bùng nổ.

*   (n!) được gọi là thời gian giai thừa. Thời gian chạy của chúng tôi hiện rất chậm, tương đương với giai thừa của kích thước đầu vào của chúng tôi.
